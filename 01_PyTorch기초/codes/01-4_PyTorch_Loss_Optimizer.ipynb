# 실제 예시

```python
x = torch.rand(10, 1)
y = 3 * x + 1
print(x)
print(y)
```

```python
model = nn.Linear(1, 1)
criterion = nn.MSELoss()        
optimizer = optim.SGD(model.parameters(), lr=0.1)

for i in range(1000):
    optimizer.zero_grad()
    pred = model(x)
    # print(type(pred), pred.size(), pred)
    loss = criterion(pred, y)
    loss.backward()
    optimizer.step()
    if (i+1)%100 == 0:
        print(i+1, "번째 Loss: ", loss.item())
```
