```python
import torch
import torch.nn as nn

print(torch.__version__)
```

# 1. 선형 회귀 예시

$output = input\cdot W + b$

```python
class myModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1) # 입력 차원 1, 출력 차원 1인 linear layer 정의

    def forward(self, x): # 순전파 정의
        return self.linear(x)
```

```python
model = myModel()
print(model)
```

```python
x = torch.tensor([1.0])
output = model.forward(x)
print(output)
```

model.forward(x)는 \__call__ 오버라이드로 model(x)가 가능하다  
=> forward를 제외한 이름들은 불가능

```python
x = torch.tensor([1.0])
output = model(x)
print(output)
```

```python
model = myModel()

for name, param in model.named_parameters(): # 파라미터 자동 수집 가능
    print(name, param.shape) 
```

# 2. MLP 예시

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(10, 20)
        self.layer2 = nn.Linear(20, 5)
        self.layer3 = nn.Linear(5, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        return self.layer3(x)
```

```python
mlpModel = MLP()
print(mlpModel)
for name, param in mlpModel.named_parameters():
    print(name, param.shape)
x = torch.randn((1, 10))  # 배치 크기 1, 입력 차원 10
print(x)
output = mlpModel(x)
print(output)
```
