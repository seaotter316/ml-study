# 1. 기본 개념 

x = 2이나, 반드시 실수 형태 -> x = 2.0 <br><br>
y에는 x → (x^2) → (x^2 + 3x) → (x^2 + 3x + 1)으로 동적 계산 그래프

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2 + 3*x + 1
print(x, y)
print(x.type(), y.type())
print(x.size(), y.size())
```

```python
y.backward() # 역전파, 미분 결과가 x.grad에 저장됨
print(x.grad)  # dy/dx
```

requires_grad=True: "이 변수의 변화에 따른 기울기를 추적해줘"

# 2. 여러 변수의 경우

```python
x = torch.tensor(1.0, requires_grad=True) # 1.0
w = torch.tensor(2.0, requires_grad=True) # 2.0
b = torch.tensor(3.0, requires_grad=True) # 3.0

y = w*x + b
y.backward()

print(x.grad)  # dy/dx
print(w.grad)  # dy/dw
print(b.grad)  # dy/db
```

# 3. 초기화

```python
for i in range(3):
    w.grad.zero_()
    y = w * x + b
    y.backward()
    print(w.grad)
```

optimizer.zero_grad() or w.grad.zero_()로 초기화

# 4. no_grad() 의미

```python
with torch.no_grad(): # 이 블록 안에서는 자동 미분이 일어나지 않음
    w -= 0.01 * w.grad
    print(w)
```
