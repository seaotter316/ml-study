## 2. 일부러 틀려 놨으니까 고쳐보쇼

```python
# 의도된 버그가 있는 간단한 CNN 모델 정의
# 여기에다가 고치지 마라
# 여기에다가 고치지 마라
# 여기에다가 고치지 마라

class BuggyCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 14 * 14, 128), 
            nn.ReLU(),
            nn.Linear(128, 10),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

buggy = BuggyCNN().to(device)
x = torch.randn(4, 1, 28, 28, device=device)

# 에러 재현
y = buggy(x)
print("output:", y.shape)
```

```python
# 여기서 고치셈

class FixedCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 7 * 7, 128), 
            nn.ReLU(),
            nn.Linear(128, 10),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
# 체크용

model = FixedCNN().to(device)
x = torch.randn(4, 1, 28, 28, device=device)

with torch.no_grad():
    y = model(x)

print("output:", y.shape)
```
